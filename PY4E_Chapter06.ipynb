```python
#Exercise 6.1: Create while loop to iterate through word from last letter to first. 

var = "pineapple"
#Take length of variable and subtract 1 to get max index, set max index as starting value
index = len(var) - 1

while index >= 0:
    letter = var[index]
    print(letter)
    index = index - 1
```
